## IMPORTING NECESSARY LIBRARIES

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

##  OBTAIN THE CURRENT WORKING DIRECTORY USING THE OS MODULE.

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "\\Data - Survey Monkey Output.xlsx", sheet_name = "Edited_Data") #load and read the excel file
dataset
```

```python
dataset_modified = dataset.copy() #make a copy of the original dataset, incase you have to revert or undo changes
dataset_modified
```

## DROP UNWANTED COLUMNS

```python
columns_to_drop = list(dataset_modified.columns[1:7])  #Selectd columns index 1 to 7 to drop
columns_to_drop 
```

```python
dataset_modified.drop(columns = columns_to_drop, inplace = True)  # Drop columns index 1 to 7
```

```python
dataset_modified.head()
```

## UNPIVOTING THE TABLE USING MELT FUNCTION

```python
#id_vars = ["Respondent ID", "Identify which division you work in. + Response", "Identify which division you work in. + Other (please specify)", 
          # "Which of the following best describes your position level? + Response", "Which generation are you apart of? + Response",
           # "Please select the gender in which you identify. + Response", "Which duration range best aligns with your tenure at your company? + Response",
           # "Which of the following best describes your employment type? + Response"]

id_vars = dataset_modified.columns[:8]
value_vars = dataset_modified.columns[8:]

dataset_melted = dataset_modified.melt(id_vars=id_vars, value_vars=value_vars, var_name= "Questions + Subquestions", value_name='Answers')

# Unpivoting the identifier columns

dataset_melted
```

```python
Questions = pd.read_excel(pwd + "\\Data - Survey Monkey Output.xlsx", sheet_name = "Questions") # read and load the questions sheet
Questions
```

## DROPPING UNWANTED COLUMNS

```python
Questions_columns_to_drop = ["Raw_Questions", "Raw_Subquestion", "Subquestions"]
Questions.drop(columns = Questions_columns_to_drop, inplace = True)
```

```python
Questions
```

## MERGING TABLES; DATASET_MODIFIED AND QUESTIONS

```python
dataset_merge = pd.merge(left = dataset_melted, right = Questions, how='left', left_on="Questions + Subquestions", right_on="Questions + Subquestions")

# Merging the two tables (Questions and main table)

print("Original dataset:", len(dataset_melted))
print("Merged dataset:", len(dataset_merge))  
dataset_merge
```

## MAKING A TABLE OF THE TOTAL RESPONDENTS PER QUESTION

```python
respondents = dataset_merge[dataset_merge["Answers"].notna()]
respondents = respondents.groupby("Questions")["Respondent ID"].nunique().reset_index()
respondents.rename(columns = {"Respondent ID" : "Respondents"}, inplace = True)
respondents
```

## MERGING TABLES; DATASET_MERGED AND RESPONDENTS

```python
dataset_merge2 = pd.merge(left = dataset_merge, right = respondents, how='left', left_on="Questions", right_on="Questions")

# Merging the two tables (respondents and dataset_merge)

print("Original dataset:", len(dataset_merge))
print("Merged dataset:", len(dataset_merge2))
dataset_merge2
```

## MAKING A TABLE OF QUESTIONS WITH SAME ANSWERS

```python
same_answer = dataset_merge   #[dataset_merge["Answers"].notna()]
same_answer = same_answer.groupby(["Questions + Subquestions", "Answers"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns = {"Respondent ID" : "same_answer"}, inplace = True)
same_answer
```

## MERGING TABLES; DATASET_MERGED2 AND SAME_ANSWERS

```python
dataset_merge3 = pd.merge(left = dataset_merge2, right = same_answer, how='left', left_on=["Questions + Subquestions", "Answers"],
                          right_on=["Questions + Subquestions", "Answers"])

# Merging the two tables (same_answer and dataset_merge2)

print("Original dataset:", len(dataset_merge2))
print("Merged dataset:", len(dataset_merge3))
dataset_merge3
```

## FILLING ALL NULL VALUES WITH 0 IN THE SAME_ANSWER COLUMN

```python
dataset_merge3["same_answer"].fillna(0, inplace = True)
```

```python
dataset_merge3
```

## RENAMING COLUMNS

```python
output = dataset_merge3.rename(columns={"Respondent ID	Identify which division you work in. + Response": "Primary Division",
"Identify which division you work in. + Other (please specify)" : "Secondary Division",
"Which of the following best describes your position level? + Response" : "Position",
"Which generation are you apart of? + Response" : "Generation",
"Please select the gender in which you identify. + Response" : "Gender",
"Which duration range best aligns with your tenure at your company? + Response" : "Tenure",
"Which of the following best describes your employment type? + Response" : "Employment Type"})
output            
```

## PLOTTING A BARCHART OF THE TOP 10 MOST ANSWERED QUESTIONS

```python
unique_questions = dataset_merge3.drop_duplicates(subset='Questions')

# Group by 'Questions' and calculate the sum of 'Respondents'
top_ten_questions_answered = unique_questions.groupby('Questions')['Respondents'].sum().nlargest(10).reset_index()

categories = top_ten_questions_answered['Questions']
values = top_ten_questions_answered['Respondents']

# Create a bar chart
plt.bar(categories, values, color='blue')

# Adding labels and title
plt.xlabel('Questions')
plt.ylabel('Number of Respondents')
plt.title('Top 10 Questions by Respondents')

# Rotate x-axis labels to be vertical
plt.xticks(rotation='vertical')

# Display the chart
plt.show()

# top_ten_questions_answered
```
